```python
# 널값 제거

null = []
for i in range(len(tag_df)):
    if len(tag_df['tag_total'][i]) < 3 :
        null.append(tag_df.index[i])
        
print(null)

tag_drop = tag_df.drop(tag_df.index[null])
tag_drop.reset_index(drop=True, inplace=True)
tag_drop
```

```python
# 중복 리스트 제거
import ast # 원래 데이터 타입으로 변환해줌

def convert_str(x):
    if type(x) is str:
        return "'"+ x + "'"
    else:
        return str(x)

df_tmp_arr = tag_drop['tag_total']
print('중복 리스트 제거 전 :', len(df_tmp_arr))
df_tmp_arr

# 단 순서가 제멋대로암 유의.
# 해석 : df_tmp_arr를 함수를 사용해 데이터형식을 바꿔주고 튜플로 출력(중복제거)
#        ast 메소드를 사용해 다시 요소들을 리스트로 바꿔주고 그 결과를 리스트로 만들어줌 
new_list = list(map(ast.literal_eval, set(map(convert_str, df_tmp_arr))))

print('중복 리스트 제거 후 :',len(new_list))
new_list
```

```python
areas = ('동', '%맛집', '안산', '대전%', '체험단')
num = 0
for j in range(len(new_list)*2):
    for i in range(len(new_list)):
        for tag in new_list[i]:
            for area in areas:
                if area in tag:
                    print(tag)
                    num += 1
                    print(num)
                    new_list[i].remove(tag)
```

```python
import re
tag_clean = []
tag_co = []
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"^0-9$"    
        u"-=+,/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》"  "]+", flags=re.UNICODE)

for i in range(len(new_list)):
    t = []
    for j in range(len(new_list[i])):
        text = new_list[i][j]
        parse = emoji_pattern.sub(r'', text) # no emoji
        if parse != '' and parse != '#':
            t.append(parse)
            tag_co.append(parse) # 빈도수때 사용
            
    tag_clean.append(t)

clean = pd.DataFrame({'tag_clean':tag_clean})
clean
```

```python
# 널값 제거
# 태그 삭제로 인해 NUll값이 발생할 수 있으므로

null = []
for i in range(len(clean)):
    if len(clean['tag_clean'][i]) < 3 :
        null.append(clean.index[i])
        
print(null)

clean = clean.drop(clean.index[null])
clean.reset_index(drop=True, inplace=True)
clean
```

```python
#### 저장 ####
list_d = pd.DataFrame(clean)
list_d.to_csv('hashtag.csv', encoding='utf-8-sig')
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_ary = te.fit(clean['tag_clean']).transform(clean['tag_clean'])
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.04, use_colnames=True)

null = []
for i in range(len(frequent_itemsets)):
    if len(frequent_itemsets['itemsets'][i]) < 2 :
        null.append(frequent_itemsets.index[i])
        
print(null)

frequent = frequent_itemsets.drop(frequent_itemsets.index[null])
 
# 항목이 2개고 지지도가 0.05이상인 것만
frequent['len'] = frequent['itemsets'].apply(lambda x: len(x))
frequent = frequent[(frequent['len']==2)].sort_values(by='support', ascending=False)
### frequent : 네트워크 분석을 위해###
frequent
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm
import numpy as np

# 그래프 생성
G = nx.DiGraph()
ar = (frequent['itemsets'])
G.add_edges_from(ar)

# 페이지 랭크 : 노드 색상과 크기 지정
pr = nx.pagerank(G)
nsize = np.array([v for v in pr.values()])
nsize = 3000 * (nsize-min(nsize)) / (max(nsize) - min(nsize))
font_name = fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

# 레이아웃
#pos = nx.circular_layout(G)
#pos = nx.shell_layout(G)
#pos = nx.kamada_kawai_layout(G)
pos = nx.random_layout(G)

# 그래프 그리기
plt.figure(figsize=(18,16))
plt.axis('off')
nx.draw_networkx(G, font_family=font_name , font_size=16, pos=pos, node_color=list(pr.values()), 
                 node_size=nsize, alpha=0.7, edge_color='.5', cmap=plt.cm.YlGn)
```

```python
result = sorted(G.degree, key=lambda x: x[1], reverse=True)
count_result = pd.DataFrame(result)
count_result[:30]
```
